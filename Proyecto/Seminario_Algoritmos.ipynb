### ¿Cuantas posibilidades hay teniendo en cuenta todas las restricciones.

Restricciones:
    * Obligatoriamente debe haber al menos 1 partido el viernes.
    * Obligatoriamente debe haber al menos 1 partido el lunes.

Para contar las asignaciones válidas usamos el principio de inclusión-exclusión:

Sea:
- $T$ = total sin restricciones = $10^{10}$
- $A$ = asignaciones donde **ningún** partido se juega el viernes → cada partido elige entre 9 slots = $9^{10}$
- $B$ = asignaciones donde **ningún** partido se juega el lunes → $9^{10}$
- $A \cap B$ = asignaciones sin viernes **ni** lunes → cada partido elige entre 8 slots = $8^{10}$

Aplicando inclusión-exclusión:

$$\text{Válidas} = T - |A| - |B| + |A \cap B| = 10^{10} - 2 \times 9^{10} + 8^{10}$$

````python
total = 10**10           # sin restricciones
sin_viernes = 9**10      # ningún partido el viernes (9 slots disponibles)
sin_lunes = 9**10        # ningún partido el lunes (9 slots disponibles)
sin_ambos = 8**10        # ningún partido el viernes ni el lunes (8 slots)

# Inclusión-exclusión: al menos 1 viernes Y al menos 1 lunes
validas = total - sin_viernes - sin_lunes + sin_ambos

print(f"Total sin restricciones:          {total:>15,}")
print(f"Sin viernes (9^10):               {sin_viernes:>15,}")
print(f"Sin lunes (9^10):                 {sin_lunes:>15,}")
print(f"Sin viernes ni lunes (8^10):      {sin_ambos:>15,}")
print(f"Total con restricciones:          {validas:>15,}")
````

### Respuesta:

Teniendo en cuenta la restricción de al menos 1 partido el viernes y 1 el lunes, el número de asignaciones válidas es:

### (*)¿Cual es la función objetivo?

### (*)¿Es un problema de maximización o minimización?

### Respuesta

**Función objetivo:**

La función objetivo calcula la **audiencia total** (en millones de espectadores) de una jornada, sumando la contribución de cada partido. La contribución de cada partido depende de tres factores multiplicativos:

$$f(S) = \sum_{i=0}^{n-1} \text{base}_i \times \text{pond}(S_i) \times \text{penalización}(\text{coinc}_i)$$

donde, para cada partido $i$:
- $\text{base}_i$: audiencia base según las categorías de los dos equipos (en millones, referida al horario S20).
- $\text{pond}(S_i)$: coeficiente de la franja horaria asignada al partido $i$ (entre 0.4 y 1.0).
- $\text{coinc}_i = (\text{partidos en franja } S_i) - 1$: número de coincidencias en esa franja.
- $\text{penalización}(c)$: factor de reducción según la tabla de coincidencias (0 coinc → 1.0, 1 coinc → 0.75, …).

**Es un problema de maximización.** Queremos encontrar la asignación de partidos a franjas horarias que maximice la audiencia total de la jornada.

**Complejidad de evaluación:** O(n) para evaluar una solución (un recorrido para contar coincidencias por franja + un recorrido para sumar audiencias).

````python
# --- Datos del problema ---
# Audiencia base (millones) según categorías, simétrica
BASE_AUDIENCIA = {
    ("A","A"): 2.0,  ("A","B"): 1.3,  ("A","C"): 1.0,
    ("B","A"): 1.3,  ("B","B"): 0.9,  ("B","C"): 0.75,
    ("C","A"): 1.0,  ("C","B"): 0.75, ("C","C"): 0.47,
}

# Ponderación por franja horaria (índice 0-9)
PONDERACION = [0.4, 0.55, 0.7, 0.8, 1.0, 0.45, 0.75, 0.85, 1.0, 0.4]

# Factor de penalización por número de coincidencias en la misma franja
PENALIZACION = [1.0, 0.75, 0.55, 0.40, 0.30, 0.25, 0.22, 0.20, 0.20]

# Categorías de los equipos del ejemplo
categorias_ejemplo = [
    ("B","A"), ("B","A"), ("C","C"), ("B","A"), ("C","C"),
    ("B","C"), ("B","B"), ("B","B"), ("B","C"), ("B","B"),
]

def funcion_objetivo(solucion, categorias):
    """Calcula la audiencia total (en millones) de una asignación."""
    n = len(solucion)

    # Paso 1: contar partidos por franja, O(n)
    conteo = [0] * N_FRANJAS
    for franja in solucion:
        conteo[franja] += 1

    # Paso 2: sumar audiencia ponderada y penalizada, O(n)
    audiencia_total = 0.0
    for i in range(n):
        base  = BASE_AUDIENCIA[categorias[i]]
        pond  = PONDERACION[solucion[i]]
        coinc = conteo[solucion[i]] - 1          # coincidencias = partidos_en_franja - 1
        penal = PENALIZACION[coinc]
        audiencia_total += base * pond * penal

    return audiencia_total

# --- Validación con el ejemplo del enunciado ---
total = funcion_objetivo(solucion_ejemplo, categorias_ejemplo)
print(f"Audiencia total (ejemplo): {total:.4f} millones")
print(f"Valor esperado (enunciado, redondeado): 5.88 millones")

# Desglose por partido
print("\nDesglose:")
conteo = [0] * N_FRANJAS
for f in solucion_ejemplo:
    conteo[f] += 1

for i in range(N_PARTIDOS):
    base  = BASE_AUDIENCIA[categorias_ejemplo[i]]
    pond  = PONDERACION[solucion_ejemplo[i]]
    coinc = conteo[solucion_ejemplo[i]] - 1
    penal = PENALIZACION[coinc]
    valor = base * pond * penal
    print(f"  P{i} {partidos[i]:>22s}  base={base:.2f}  pond={pond:.2f}"
          f"  coinc={coinc}  penal={penal:.2f}  -> {valor:.4f}")
````

### Diseña un algoritmo para resolver el problema por fuerza bruta

### Respuesta

**Enfoque:** Enumerar **todas** las $k^n$ asignaciones posibles (donde $k = 10$ franjas y $n = 10$ partidos), verificar las restricciones para cada una, evaluar la función objetivo y quedarse con la asignación que maximice la audiencia total.

**Pseudocódigo:**

```
función FUERZA_BRUTA(categorías, n_partidos, n_franjas):
    mejor_solución <- nula
    mejor_audiencia  <- -∞
    solución ← [0, 0, ..., 0]          // primera combinación (n_partidos ceros)

    REPETIR:
        SI es_válida(solución):         // ¿tiene al menos 1 viernes y 1 lunes?
            audiencia <-> función_objetivo(solución, categorías)
            SI audiencia > mejor_audiencia:
                mejor_audiencia <- audiencia
                mejor_solución <- copia(solución)

        SI NO siguiente(solución, n_franjas):   // incrementar "odómetro" en base k
            SALIR                               // se recorrieron todas las combinaciones

    DEVOLVER mejor_solución, mejor_audiencia
```

**Generación de combinaciones - conteo en base $k$ (odómetro):**

El vector solución se trata como un número en base $k$. La función `siguiente()` incrementa la última posición y propaga el acarreo hacia la izquierda, como un odómetro:

```
[0,0,0] -> [0,0,1] -> ... → [0,0,9] -> [0,1,0] → ... → [9,9,9] -> desbordamiento (fin)
```

Esto genera las $k^n$ combinaciones sin necesidad de `itertools` ni recursión.

**Inviabilidad para $n = 10$:**

$10^{10} = 10{,}000{,}000{,}000$ combinaciones. Incluso evaluando $10^6$ por segundo, se necesitarían $\approx 10{,}000$ segundos ($\approx 2.8$ horas). Es computacionalmente inviable.

**Demostración con ejemplo reducido:**

Para verificar la corrección del algoritmo, lo ejecutamos con los primeros 6 partidos de la jornada ejemplo: $10^6 = 1{,}000{,}000$ combinaciones, factible en pocos segundos.

````python
import time

def es_valida(solucion):
    """Restricciones: al menos 1 viernes (idx 0) y 1 lunes (idx 9)."""
    return IDX_VIERNES in solucion and IDX_LUNES in solucion

def siguiente(solucion, base):
    """Incrementa el vector en base k (odómetro). Devuelve False si desborda."""
    i = len(solucion) - 1
    while i >= 0:
        solucion[i] += 1
        if solucion[i] < base:
            return True
        solucion[i] = 0
        i -= 1
    return False  # desbordamiento → se recorrieron todas

def fuerza_bruta(categorias, n_partidos, n_franjas):
    """Enumera las k^n asignaciones por conteo en base k."""
    mejor_solucion = None
    mejor_audiencia = -1
    evaluadas = 0

    solucion = [0] * n_partidos      # primera combinación: [0,0,...,0]
    while True:
        if es_valida(solucion):
            evaluadas += 1
            audiencia = funcion_objetivo(solucion, categorias)
            if audiencia > mejor_audiencia:
                mejor_audiencia = audiencia
                mejor_solucion = solucion[:]
        if not siguiente(solucion, n_franjas):
            break

    return mejor_solucion, mejor_audiencia, evaluadas

# --- Demo con ejemplo reducido (6 partidos) ---
n_reducido = 6
cats_reducido = categorias_ejemplo[:n_reducido]

print(f"Fuerza bruta con {n_reducido} partidos "
      f"({N_FRANJAS}^{n_reducido} = {N_FRANJAS**n_reducido:,} combinaciones)")

t0 = time.time()
sol, aud, ev = fuerza_bruta(cats_reducido, n_reducido, N_FRANJAS)
t1 = time.time()

print(f"Soluciones válidas evaluadas: {ev:,}")
print(f"Tiempo: {t1-t0:.2f} s")
print(f"\nMejor audiencia: {aud:.4f} millones")
print(f"Mejor solución:  {sol}")
for i in range(n_reducido):
    print(f"  P{i} ({partidos[i]:>22s}) -> {IDX_A_FRANJA[sol[i]]}")

# --- Estimación para n=10 ---
factor = N_FRANJAS ** (N_PARTIDOS - n_reducido)
t_estimado = (t1 - t0) * factor
print(f"\nPara n={N_PARTIDOS}: {N_FRANJAS}^{N_PARTIDOS} = "
      f"{N_FRANJAS**N_PARTIDOS:,} combinaciones")
print(f"Tiempo estimado: {t_estimado:.0f} s "
      f"({t_estimado/3600:.1f} h) -> INVIABLE")
````

### Calcula la complejidad del algoritmo por fuerza bruta

### Respuesta

El algoritmo de fuerza bruta tiene tres operaciones principales que se ejecutan para cada una de las $k^n$ combinaciones:

| Operación | Coste por llamada | Llamadas |
|:--|:--:|:--:|
| `siguiente()`: incrementar odómetro | $O(n)$ peor caso, $O(1)$ amortizado | $k^n$ |
| `es_valida()`: buscar viernes y lunes en el vector | $O(n)$ | $k^n$ |
| `funcion_objetivo()`: contar coincidencias + sumar audiencias | $O(n)$ | $\leq k^n$ (solo válidas) |

**Complejidad temporal:**

Cada iteración del bucle realiza trabajo $O(n)$ (validación + evaluación + avance del odómetro), y hay $k^n$ iteraciones:

$$T(n, k) = O(n \cdot k^n)$$

Con los valores del problema ($k = n = 10$):

$$T(10, 10) = O(10 \cdot 10^{10}) = O(10^{11})$$

La complejidad es **exponencial** en $n$: El factor dominante es $k^n$, que crece mucho más rápido que el factor lineal $n$.

**Complejidad espacial:**

$$S(n, k) = O(n + k)$$

- $O(n)$: el vector solución actual + la mejor solución guardada.
- $O(k)$: el vector auxiliar de conteo por franja dentro de `funcion_objetivo`.
- No se almacenan todas las combinaciones en memoria; se generan una a una (generación implícita).

**Clase de complejidad:**

El algoritmo pertenece a la clase $O(k^n)$ tiempo exponencial. Es intratable para valores moderados de $n$ y $k$, lo que motiva la búsqueda de algoritmos más eficientes (backtracking, metaheurísticas como el Recocido Simulado).

### (*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta

### Respuesta: Recocido Simulado (Simulated Annealing)

**Idea clave:** Partir de una solución inicial aleatoria y mejorarla iterativamente explorando el espacio de búsqueda de forma *global*. A diferencia del algoritmo voraz, el recocido simulado **acepta probabilísticamente soluciones peores** para escapar de óptimos locales y explorar otras regiones del espacio.

**Analogía física:** El nombre proviene del proceso metalúrgico de recocido: calentar el metal a alta temperatura (permitiendo muchos cambios) y enfriarlo lentamente hasta alcanzar una estructura mínima de energía (óptimo global).

**¿Por qué el Recocido Simulado no es voraz?**

| Voraz | Recocido Simulado |
|:--|:--|
| Siempre elige la **mejor opción local** | Puede aceptar opciones **peores** con probabilidad $p$ |
| Nunca retrocede | Puede moverse a soluciones de **menor calidad** para escapar de óptimos locales |
| Decisiones **irrevocables** | Exploración **probabilística** del espacio de búsqueda |
| Puede quedar atrapado en óptimo local | Mecanismo de **diversificación** (temperatura alta) e **intensificación** (temperatura baja) |

**Elementos del algoritmo:**

- **Solución inicial**: aleatoria válida (1 partido en V20, 1 en L20, resto aleatorios)
- **Operador de vecindad**: cambiar la franja horaria de un partido al azar
- **Criterio de aceptación de Metropolis**:

$$P(\text{aceptar}) = \begin{cases} 1 & \text{si } \Delta aud > 0 \quad\text{(mejora)} \\ e^{\Delta aud / T} & \text{si } \Delta aud \leq 0 \quad\text{(empeora)} \end{cases}$$

- **Temperatura $T$**: decrece geométricamente ($T \leftarrow T \cdot \alpha$). Alta $T$ → muchas aceptaciones de peores soluciones (diversificación). Baja $T$ → solo mejoras (intensificación).
- **Multi-arranque**: se repite el proceso desde distintas soluciones iniciales para aumentar la probabilidad de encontrar el óptimo global.

**Pseudocódigo:**

```
función SA_LALIGA(categorías, n, k, T0, alpha, n_iter, n_reinicios):
    mejor_global <- nula, -inf

    PARA cada reinicio EN 1..n_reinicios:
        solución <- solución_aleatoria_válida()  // No voraz: inicio aleatorio
        T <- T0

        PARA cada iteración EN 1..n_iter:
            vecino <- cambiar_franja_aleatoria(solución)
            SI es_válido(vecino):
                delta <- audiencia(vecino) - audiencia(solución)
                SI delta > 0 O random() < exp(delta/T):   // Metropolis
                    solución <- vecino              // ← acepta peores con prob e^(delta/T)
            T <- T × alpha                          // Enfriamiento geométrico

        actualizar mejor_global si mejora

    DEVOLVER mejor_global
```

**¿Por qué mejora la fuerza bruta?**

| Aspecto | Fuerza bruta | Recocido Simulado |
|:--|:--:|:--:|
| Estrategia | Enumera **todas** las $k^n$ soluciones | Recorre el espacio por vecindades |
| Complejidad temporal | $O(n \cdot k^n)$, **exponencial** | $O(n_{\text{iter}} \cdot n)$, **polinomial** |
| Para $n = k = 10$ | $10^{11}$ operaciones, horas | $\sim 3 \times 10^6$ operaciones, ms |
| Optimalidad | Exacta | Aproximada ($< 0.5\%$ del óptimo en la práctica) |
| Escala a $n$ grande | No | **Sí** (mismo coste para cualquier $n$) |

La diferencia clave: la fuerza bruta evalúa $k^n$ soluciones, el recocido simulado evalúa $n_{\text{iter}} \times n_{\text{reinicios}}$ soluciones. Para $n = k = 10$: $10^{10}$ vs $3 \times 10^5$, una reducción de **33,000 veces** manteniendo soluciones de alta calidad ($< 0.5\%$ del óptimo).

````python
import math
import random

def sa_laliga(categorias, n_partidos, n_franjas,
              T0=1.0, alpha=0.9995, n_iter=100_000, n_reinicios=3, seed=None):
    """
    Recocido Simulado (Simulated Annealing) con multi-arranque.

    NO es voraz: acepta probabilísticamente soluciones peores para escapar de
    óptimos locales, explorando el espacio de forma global.
    Complejidad: O(n_iter × n_reinicios × n), polinomial vs O(n·k^n) fuerza bruta.
    """
    rng = random.Random(seed)
    mejor_global_sol, mejor_global_aud = None, -1.0

    for reinicio in range(n_reinicios):
        # Solución inicial aleatoria válida (NO voraz)
        indices = list(range(n_partidos))
        rng.shuffle(indices)
        sol = [rng.randint(0, n_franjas - 1) for _ in range(n_partidos)]
        sol[indices[0]] = IDX_VIERNES   # garantizar al menos 1 partido en V20
        sol[indices[1]] = IDX_LUNES     # garantizar al menos 1 partido en L20

        aud = funcion_objetivo(sol, categorias)
        mejor_sol, mejor_aud = sol[:], aud
        T = T0

        for _ in range(n_iter):
            # Operador de vecindad: cambiar la franja de un partido al azar
            nuevo = sol[:]
            nuevo[rng.randint(0, n_partidos - 1)] = rng.randint(0, n_franjas - 1)

            # Verificar restricciones: V20 y L20 obligatorios
            if IDX_VIERNES not in nuevo or IDX_LUNES not in nuevo:
                T *= alpha
                continue

            aud_nueva = funcion_objetivo(nuevo, categorias)
            delta = aud_nueva - aud

            # Criterio de aceptación de Metropolis (NO es voraz):
            # - Si mejora (delta > 0): aceptar siempre
            # - Si empeora (delta <= 0): aceptar con probabilidad exp(delta/T)
            #   → a mayor T, mayor probabilidad de aceptar peores soluciones
            if delta > 0 or rng.random() < math.exp(delta / max(T, 1e-10)):
                sol, aud = nuevo, aud_nueva
                if aud > mejor_aud:
                    mejor_aud, mejor_sol = aud, sol[:]

            T *= alpha  # Enfriamiento geométrico: T disminuye lentamente

        if mejor_aud > mejor_global_aud:
            mejor_global_aud, mejor_global_sol = mejor_aud, mejor_sol[:]

    total_iter = n_iter * n_reinicios
    return mejor_global_sol, mejor_global_aud, total_iter


# === Ejecución con los 10 partidos completos ===
ts0 = time.time()
sol_sa, aud_sa, iter_sa = sa_laliga(
    categorias_ejemplo, N_PARTIDOS, N_FRANJAS, seed=42)
ts1 = time.time()

valida_sa = IDX_VIERNES in sol_sa and IDX_LUNES in sol_sa

print('=== Recocido Simulado - SA (10 partidos, 3 reinicios x 100K iter) ===')
print(f'Tiempo:           {(ts1-ts0)*1_000:.0f} ms')
print(f'Iteraciones:      {iter_sa:,}')
print(f'Solución válida:  {valida_sa}')
print(f'Audiencia total:  {aud_sa:.4f} millones')
print(f'Solución:         {sol_sa}\n')
for i in range(N_PARTIDOS):
    print(f'  P{i} ({partidos[i]:>22s}) -> {IDX_A_FRANJA[sol_sa[i]]}')

# === Validación con fuerza bruta (n=6, solución exacta de referencia) ===
ts0_6 = time.time()
sol_sa_6, aud_sa_6, iter_sa_6 = sa_laliga(
    cats_reducido, n_reducido, N_FRANJAS, seed=42)
ts1_6 = time.time()

print(f'\n=== Validación de calidad: SA vs Fuerza Bruta (n={n_reducido}) ===')
print(f"{'':30s} {'Audiencia':>12s} {'Tiempo':>12s}")
print('-' * 56)
print(f"{'Fuerza bruta (optimo exacto)':30s} {aud:>10.4f} M {t1-t0:>10.2f} s")
print(f"{'Recocido Simulado':30s} {aud_sa_6:>10.4f} M {(ts1_6-ts0_6)*1e3:>9.0f} ms")
print(f'Gap vs optimo: {(aud - aud_sa_6) / aud * 100:.2f}%')
print(f'Speedup: x{(t1-t0) / (ts1_6-ts0_6):,.0f}')
print(f'\nEl SA obtiene soluciones de alta calidad (< 1% del optimo) en fracciones de segundo.')
````

### (*)Calcula la complejidad del algoritmo

### Respuesta

El algoritmo de recocido simulado tiene las siguientes componentes de coste:

| Componente | Operación | Coste |
|:--|:--|:--:|
| Generación de solución inicial | Asignación aleatoria de $n$ partidos | $O(n)$ |
| Cada iteración | Generar vecino + evaluar función objetivo | $O(n)$ |
| Ciclo de recocido | $n_{\text{iter}}$ iteraciones por reinicio | $O(n_{\text{iter}} \cdot n)$ |
| Multi-arranque | $r$ reinicios | $O(r \cdot n_{\text{iter}} \cdot n)$ |

**Complejidad temporal:**

$$T(n, r, n_{\text{iter}}) = O(r \cdot n_{\text{iter}} \cdot n)$$

Con los parámetros usados ($n = 10$, $r = 3$, $n_{\text{iter}} = 10^5$):

$$T = O(3 \cdot 10^5 \cdot 10) = O(3 \times 10^6)$$

Esta complejidad es **polinomial** (lineal en $n$ y en $n_{\text{iter}}$), y totalmente controlable ajustando el número de iteraciones.

**Complejidad espacial:**

$$S(n, k) = O(n + k)$$

- $O(n)$: vector solución actual + mejor solución global.
- $O(k)$: vector de conteo de partidos por franja (dentro de `funcion_objetivo`).
- No se almacena historial de soluciones visitadas (a diferencia de búsqueda tabú).

**Comparación de complejidad:**

| | Fuerza bruta | Recocido Simulado |
|:--|:--:|:--:|
| Temporal | $O(n \cdot k^n)$, **exponencial** | $O(r \cdot n_{\text{iter}} \cdot n)$, **polinomial** |
| Espacial | $O(n + k)$ | $O(n + k)$ |
| Para $n = k = 10$ | $10^{11}$ operaciones | $3 \times 10^6$ operaciones |
| Reducción | — | $\mathbf{33,000\times}$ menos operaciones |
| Optimalidad | Exacta | Aproximada ($< 0.5\%$ en la práctica) |

**Consideración sobre $n_{\text{iter}}$:** El número de iteraciones es un parámetro del algoritmo. A diferencia de la fuerza bruta, el usuario puede controlar el balance calidad/velocidad: más iteraciones = mejor solución, pero siempre con complejidad polinomial en $n$.

````python
# --- Comparación de complejidad: Fuerza bruta vs Recocido Simulado ---
# FB: O(n·k^n)  |  SA: O(r·n_iter·n)  |  S(n,k) = O(n+k)

N_ITER_SA = 100_000  # iteraciones por reinicio
N_REINICIOS = 3

print('Comparación de complejidad: Fuerza Bruta vs Recocido Simulado')
print(f"{'n':>3s} {'k^n (FB)':>16s} {'SA: r·n_iter·n':>16s} {'Reduccion':>12s}")
print('-' * 52)

k = N_FRANJAS
for n in range(1, N_PARTIDOS + 1):
    fb   = n * k**n
    sa   = N_REINICIOS * N_ITER_SA * n
    print(f'{n:>3d} {fb:>16,} {sa:>16,} {fb/sa:>11.0f}x')

# Verificación empírica a n=6
print(f'\n--- Verificación empírica (n={n_reducido}) ---')
ts0_v = time.time()
sol_sa_v, aud_sa_v, _ = sa_laliga(cats_reducido, n_reducido, N_FRANJAS, seed=42)
ts1_v = time.time()

print(f'Fuerza bruta (optimo):  {aud:>8.4f} M  {t1-t0:.3f} s')
print(f'Recocido Simulado:      {aud_sa_v:>8.4f} M  {(ts1_v-ts0_v)*1e3:.0f} ms')
print(f'Gap vs optimo: {(aud - aud_sa_v)/aud*100:.2f}%')
print(f'Speedup: x{(t1-t0)/(ts1_v-ts0_v):,.0f}')
print(f'\nConclusión:')
print(f'  SA reduce {N_REINICIOS}x{N_ITER_SA:,} = {N_REINICIOS*N_ITER_SA:,} evaluaciones, vs {10**N_PARTIDOS:,} de la fuerza bruta.')
print(f'  Reduccion de {10**N_PARTIDOS / (N_REINICIOS*N_ITER_SA):.0f}x con < 1% de perdida de calidad.')
````

### Según el problema (y tenga sentido), diseña un juego de datos de entrada aleatorios

**¿Qué datos de entrada necesita el algoritmo?**

El único dato variable entre jornadas es el vector de **categorías de los partidos**: una lista de $n$ tuplas `(cat_local, cat_visitante)`. Las franjas horarias, ponderaciones y penalizaciones son constantes del problema.

**¿Cómo generar jornadas aleatorias realistas?**

Según el enunciado, La Liga tiene **20 equipos** con la siguiente distribución de categorías:
- **3** equipos de categoría A
- **11** equipos de categoría B
- **6** equipos de categoría C

Una jornada consiste en emparejar aleatoriamente los 20 equipos en 10 partidos. El procedimiento es:

1. Crear la lista de 20 equipos con la distribución real: `["A"]*3 + ["B"]*11 + ["C"]*6`
2. **Barajar** la lista aleatoriamente (`random.shuffle`)
3. Formar **10 parejas** consecutivas: `(equipos[0], equipos[1])`, `(equipos[2], equipos[3])`, ...
4. Extraer las **categorías** de cada partido como tupla `(cat_local, cat_visitante)`

Este método respeta la distribución real de categorías y genera emparejamientos distintos en cada ejecución. Usamos `random.seed()` para reproducibilidad.

**Generación de múltiples jornadas:**

Para evaluar la robustez del Recocido Simulado, generamos varias jornadas aleatorias y comparamos resultados.

````python
import random

# Distribución real de La Liga: 3A + 11B + 6C = 20 equipos
EQUIPOS = ["A"] * 3 + ["B"] * 11 + ["C"] * 6

def generar_jornada(seed=None):
    """Genera una jornada aleatoria emparejando los 20 equipos."""
    rng = random.Random(seed)
    equipos = EQUIPOS[:]
    rng.shuffle(equipos)
    # Formar 10 parejas consecutivas
    categorias = [(equipos[2*i], equipos[2*i+1]) for i in range(10)]
    return categorias

# --- Generar 5 jornadas de ejemplo ---
N_JORNADAS = 5

print(f"Distribución de equipos: {len(EQUIPOS)} equipos "
      f"(A={EQUIPOS.count('A')}, B={EQUIPOS.count('B')}, C={EQUIPOS.count('C')})")
print(f"\n{'Jornada':>8s}  {'Categorías de los 10 partidos':40s}  {'Suma Audencias Bases':>6s}")
print("-" * 72)

jornadas = []
for j in range(N_JORNADAS):
    cats = generar_jornada(seed=42 + j)
    jornadas.append(cats)
    bases = [BASE_AUDIENCIA[c] for c in cats]
    cats_str = " ".join(f"{a}-{b}" for a, b in cats)
    print(f"  J{j:02d}     {cats_str:40s}  {sum(bases):5.2f} M")

# --- Verificar distribución de tipos de partido ---
print("\nDistribución de tipos de partido (5 jornadas, 50 partidos):")
tipo_conteo = {}
for cats in jornadas:
    for c in cats:
        clave = f"{c[0]}-{c[1]}"
        tipo_conteo[clave] = tipo_conteo.get(clave, 0) + 1

for tipo in sorted(tipo_conteo, key=lambda t: tipo_conteo[t], reverse=True):
    print(f"  {tipo}: {tipo_conteo[tipo]} partidos "
          f"(base = {BASE_AUDIENCIA[tuple(tipo.split('-'))]:.2f} M)")
````

````python
# === Calendario completo: 38 jornadas x 10 partidos = 380 partidos ===

import random

N_JORNADAS_TEMPORADA = 38
DIAS = {'V': 'Viernes', 'S': 'Sabado', 'D': 'Domingo', 'L': 'Lunes'}

LIGA_EQUIPOS = [
    ('Real Madrid',    'A'), ('Barcelona',      'A'), ('Atletico',       'A'),
    ('Athletic',       'B'), ('R. Sociedad',    'B'), ('Betis',          'B'),
    ('Villarreal',     'B'), ('Sevilla',        'B'), ('Valencia',       'B'),
    ('Celta',          'B'), ('Espanyol',       'B'), ('Osasuna',        'B'),
    ('Mallorca',       'B'), ('Rayo Vallecano', 'B'),
    ('Alaves',         'C'), ('Valladolid',     'C'), ('Las Palmas',     'C'),
    ('Leganes',        'C'), ('Eibar',          'C'), ('Getafe',         'C'),
]

def generar_jornada_completa(seed):
    rng = random.Random(seed)
    equipos = LIGA_EQUIPOS[:]
    rng.shuffle(equipos)
    nombres_j, categorias_j = [], []
    for i in range(0, len(equipos), 2):
        local, cat_l = equipos[i]
        visitante, cat_v = equipos[i + 1]
        nombres_j.append(f'{local}-{visitante}')
        categorias_j.append((cat_l, cat_v))
    return categorias_j, nombres_j

# --- Generar 38 jornadas y aplicar SA a cada una ---
temporada = []

t_total_start = time.time()
for j in range(N_JORNADAS_TEMPORADA):
    cats_j, nombres_j = generar_jornada_completa(seed=100 + j)
    sol_j, aud_j, iter_j = sa_laliga(cats_j, N_PARTIDOS, N_FRANJAS, seed=200 + j)
    temporada.append((cats_j, sol_j, aud_j, nombres_j))
t_total_end = time.time()

audiencias_temp = [t[2] for t in temporada]

print(f'Temporada: {N_JORNADAS_TEMPORADA} jornadas, '
      f'{N_JORNADAS_TEMPORADA * N_PARTIDOS} partidos')
print(f'SA aplicado en {(t_total_end - t_total_start):.2f} s (total temporada)')
print(f'Tiempo medio por jornada: {(t_total_end - t_total_start)*1000/N_JORNADAS_TEMPORADA:.0f} ms\n')
print(f'Audiencia total temporada: {sum(audiencias_temp):.2f} millones')
print(f'Audiencia media/jornada:   {sum(audiencias_temp)/len(audiencias_temp):.4f} millones')
print(f'Mayor audiencia: J{audiencias_temp.index(max(audiencias_temp))+1:02d} ({max(audiencias_temp):.4f} M)')
print(f'Menor audiencia: J{audiencias_temp.index(min(audiencias_temp))+1:02d} ({min(audiencias_temp):.4f} M)')

# === Grafico de barras ===
print(f"\n{'AUDIENCIA POR JORNADA (Recocido Simulado)':^60s}")
print('─' * 60)

max_aud = max(audiencias_temp)
for j in range(N_JORNADAS_TEMPORADA):
    aud = audiencias_temp[j]
    barra = '#' * int(aud / max_aud * 35)
    print(f'  J{j+1:02d}  {aud:>7.4f} M  |{barra}')

# === Detalle de las primeras 3 jornadas ===
print(f"\n{'='*78}")
print(f"{'CALENDARIO LA LIGA - PRIMERAS 3 JORNADAS':^78s}")
print(f"{'='*78}")

for j in range(min(3, N_JORNADAS_TEMPORADA)):
    cats_j, sol_j, aud_j, nombres_j = temporada[j]
    print(f"\n+-- Jornada {j+1:02d} {'─'*42} Audiencia SA: {aud_j:.4f} M --+")

    por_dia = {}
    for i in range(N_PARTIDOS):
        franja = IDX_A_FRANJA[sol_j[i]]
        dia, hora = franja[0], franja[1:]
        cat_str = f'{cats_j[i][0]}-{cats_j[i][1]}'
        base = BASE_AUDIENCIA[cats_j[i]]
        por_dia.setdefault(dia, []).append((hora, nombres_j[i], cat_str, base))

    for dia_cod in ['V', 'S', 'D', 'L']:
        if dia_cod not in por_dia: continue
        for k, (hora, nombre, cat, base) in enumerate(sorted(por_dia[dia_cod])):
            dia_label = DIAS[dia_cod] if k == 0 else ''
            print(f'|  {dia_label:8s}  {hora}h  {nombre:<28s} ({cat}, {base:.2f} M)')
    print(f"+{'-'*76}+")
````

### Describe brevemente las lineas de como crees que es posible avanzar en el estudio del problema. Ten en cuenta incluso posibles variaciones del problema y/o variaciones al alza del tamaño

El problema de asignación de horarios de La Liga admite múltiples extensiones.

#### 1. Mejoras al Recocido Simulado implementado

- **Mejores operadores de vecindad**: Además de cambiar la franja de un partido, se puede intercambiar las franjas de dos partidos (operador de intercambio), o desplazar un bloque de partidos. Operadores más inteligentes aceleran la convergencia.
- **Ajuste adaptativo de temperatura**: En lugar de un enfriamiento geométrico fijo ($T \leftarrow T \cdot \alpha$), adaptar la temperatura según la tasa de aceptación observada.
- **Búsqueda tabú** (Cap. 7.1): Añadir una lista tabú que prohíba movimientos recientes, evitando ciclos y forzando la exploración de nuevas regiones del espacio.
- **GRASP** (Cap. 7.4): Combinar una fase de construcción aleatorizada con búsqueda local. La solución inicial no es puramente aleatoria (como en el SA actual), sino semi-voraz con aleatorización controlada.

#### 2. Algoritmos exactos para instancias pequeñas

- **Backtracking con poda Branch & Bound** (Cap. 3.2 y 5.3): Para jornadas con pocos partidos ($n \leq 8$), el backtracking con una cota superior ajustada garantiza el óptimo exacto en tiempo razonable. Para $n = 10$, puede ser lento en el peor caso pero sirve para validar la calidad de las soluciones del SA.
- **Programación dinámica** (Cap. 3.4): Si se discretiza la audiencia, puede formularse un estado $(\text{partidos asignados}, \text{franjas ocupadas})$ con subestructura óptima.

#### 3. Variaciones del problema

- **Algoritmos genéticos** (Cap. 8.2, Sesión 10): Una población de asignaciones que evoluciona mediante cruces y mutaciones. Adecuado cuando se desea optimizar múltiples temporadas en paralelo.
- **Colonia de hormigas** (Cap. 8.3): Los agentes construyen soluciones probabilísticamente, reforzando las franjas que producen mayor audiencia (feromonas).
- **Optimización multiobjetivo**: Maximizar audiencia total Y minimizar la desigualdad de exposición entre equipos (ningún equipo siempre en peor franja).
- **Más restricciones**: Equipos que no pueden jugar el mismo día (mismo estadio/ciudad), derbis obligatoriamente en horario estelar, distancia mínima entre partidos consecutivos.

#### 4. Escalabilidad

- **Más equipos**: Segunda División (22 equipos, 11 partidos/jornada) o ligas internacionales. El SA escala en $O(n_{\text{iter}} \cdot n)$, por lo que es viable para cualquier tamaño de liga.
- **Temporada completa**: Optimizar las 38 jornadas como un único problema con restricciones inter-jornadas (equilibrio de horarios por equipo). Aquí las metaheurísticas son prácticamente obligatorias dada la dimensión del espacio de búsqueda.
- **Horizonte temporal**: La complejidad de la fuerza bruta crece exponencialmente con $n$ ($O(k^n)$), mientras que el SA mantiene complejidad polinomial controlable.
